# Word level Writer Identification

## Reading Dataset IAM

IAM dataset have ASCII file with all the details for each form. We parse the form.txt to get Writer id and their corresponding forms.


We created a dictionary where key is writers'id and value is a list of form_id written by the writer.

```python
# Create a dictionary to store each form ID and its writer
from itertools import islice
from collections import defaultdict
```

```python
form_writer = defaultdict(list)
forms_file_path = "./data/forms.txt"
with open(forms_file_path) as f:
    for line in islice(f, 16, None):
        line_list = line.split(' ')
        form_id = line_list[0][:7]
        writer = line_list[1]
        form_writer[form_id].append(writer)
# list(form_writer.items())
```

```python
print("Form_id \t No. of Writer")
no_of_form_no_of_writer = defaultdict(int)
for key, value in sorted(form_writer.items(), key= lambda kv : len(kv[1]),reverse= True):
    print(f"{key}\t\t\t{len(value)}")
    no_of_form_no_of_writer[len(value)] += 1
```
